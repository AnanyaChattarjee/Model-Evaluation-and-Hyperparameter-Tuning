```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_breast_cancer
```

DATA LOADING:

```python
breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target
```

```python
data1 = np.c_[breast_cancer.data, breast_cancer.target]
```

```python
columns = np.append(breast_cancer.feature_names,["target"])
```

```python
data = pd.DataFrame(data1, columns=columns)
```

```python
X = data
```

```python
y = data["target"]
```

```python
X.drop("target",axis=1,inplace=True)
```

DATA SPLITTING:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
X_train
```

```python
models = {
    "Logistic Regression": {
        "estimator": LogisticRegression(max_iter=1000),
        "param_grid": {
            'C': [0.01, 0.1, 1, 10],
            'solver': ['liblinear', 'lbfgs']
        }
    },
    "Random Forest": {
        "estimator": RandomForestClassifier(),
        "param_grid": {
            'n_estimators': [100, 200, 300],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5]
        }
    },
    "SVM": {
        "estimator": SVC(),
        "param_grid": {
            'C': [0.1, 1, 10],
            'kernel': ['linear', 'rbf'],
            'gamma': ['scale', 'auto']
        }
    }
}
```

. we have created a dictionary named models that contains models we want to train and it's grid parameters so that we can perform looping.

```python
results = []

for name, config in models.items():
    print(f"\n Tuning {name}...")

    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('model', config['estimator'])
    ])

    # GridSearchCV
    grid = GridSearchCV(pipeline, {'model__' + k: v for k, v in config['param_grid'].items()},
                        cv=5, scoring='f1', n_jobs=-1)
    grid.fit(X_train, y_train)
    y_pred_grid = grid.predict(X_test)
    f1_grid = f1_score(y_test, y_pred_grid)

    # RandomizedSearchCV
    random = RandomizedSearchCV(pipeline, {'model__' + k: v for k, v in config['param_grid'].items()},
                                cv=5, scoring='f1', n_iter=10, random_state=42, n_jobs=-1)
    random.fit(X_train, y_train)
    y_pred_rand = random.predict(X_test)
    f1_rand = f1_score(y_test, y_pred_rand)

    results.append({
        "Model": name,
        "F1 GridSearch": f1_grid,
        "Best Params (GridSearch)": grid.best_params_,
        "F1 RandomSearch": f1_rand,
        "Best Params (RandomSearch)": random.best_params_
    })

# Display Results
df_results = pd.DataFrame(results)
print("\n Final Results:")
print(df_results)

# Best model selection
best_model_row = df_results.loc[df_results[['F1 GridSearch', 'F1 RandomSearch']].max(axis=1).idxmax()]
print("\n Best Model Overall:")
print(best_model_row)
```

. we have created a list named results in which we'll be saving all our f1 scores.
. we are looping over the models dictionary to one by one select one model and calculate it's metrices.
. further performing grid search cv to get the best performing parameters.
. and the same for random search cv to get the best parameters in the fastest manner.
. we can easily conclude that the logistic regression is definitly performing better than random forest or SVM.
. hence it returns logistic regression for the best model which we calculated by getting the max f1 score for both grid search and random search.
